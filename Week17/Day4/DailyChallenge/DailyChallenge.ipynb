```python
import matplotlib.pyplot as plt
import numpy as np   
import pandas as pd
```

```python
x = np.linspace(0,20,100)
y = x ** 2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x vs y (y = x ^ 2)')
plt.plot(x,y)
plt.show()
```

```python
chipo = pd.read_csv('./chipo.csv')
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace("$", "")))
x = chipo['item_price']
y = chipo['quantity']

plt.xlabel('Item price')
plt.ylabel('Quantity ordered')
plt.title('Relationship between price and quantity ordered')
plt.scatter(x,y, marker=".")
plt.savefig('price against quantity.png')
plt.show()
```

```python
items = chipo.groupby('item_name').sum()['quantity'].sort_values(axis=0, ascending=False).head(5)
x = items.index.values
y = list(items.values)
fig, ax = plt.subplots()
plt.xlabel('Item names')
plt.ylabel('Quantity ordered')
plt.title('Top 5 items by quantity ordered')
ax.tick_params(axis='x', labelrotation=75)
ax.bar(x,y)
fig.show()
```
