```python
x = np.arange(0,20,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
y = np.cos(x)
plt.plot(x,y, color='green', marker=".")
```

```python
x = np.arange(0,20,0.1)
y = np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,y)
y = np.cos(x)
plt.subplot(2,1,2)
plt.plot(x,y)
```

```python
x = np.arange(0,20,0.1)
y = np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,y)
plt.legend(['y = sin(x)'], loc='upper right')
plt.show()
y = np.cos(x)
plt.subplot(2,1,2)
plt.plot(x,y)
plt.legend(['y = cos(x)'], loc='upper right')
plt.show()
```
