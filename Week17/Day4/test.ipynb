```python
import pandas as pd
import numpy as np
import matplotlib as plt

df = pd.read_csv('./sales.csv')
df = df.dropna()
df.head(10)
```

```python
df.plot(kind='scatter', x='Rating', y='Unit price', color='green')
plt.pyplot.show() # type: ignore
```

```python
print(df['Gender'].unique())
```

```python
df_gender = df.groupby('Gender').count()['Invoice ID']
df_gender.plot(kind='bar', xlabel='Gender', ylabel='Number of sales', title='Sales by gender')
```

```python
df_city = df.groupby('City').count()['Invoice ID']
# plot a pie chart using df.plot()
df_city.plot(kind='pie', y='Invoice ID', title='Sales by City')
```

```python
df_city.plot(kind='bar',x='City' ,y='Invoice ID', title='Sales by City')
```

```python
sales_by_city = df.groupby('City').sum()[['Total', 'gross income']]
# plot to bar chart using df.plot()
sales_by_city.plot(kind='bar')
```

```python
profit_by_city = df[['City', 'gross margin percentage', 'Tax 5%']]
profit_by_city = profit_by_city.groupby('City').mean()
profit_by_city.plot(kind='bar', title='Average profit and tax by city')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)
y = x**2
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.annotate('Intersection', xy=(1,1), xytext=(20,20),arrowprops=dict(facecolor='black'))
plt.title('Square and Square Root of x')
plt.plot(x,y)
plt.plot(y,x,color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(100) *10
y = np.random.randn(100) * 100

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')

figure, axes = plt.subplots()
axes.scatter(x,y)
figure.savefig('Scatter diagram.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y1 = x**2
y2 = x**3

fig, axes = plt.subplots()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of x^2 and x^3')
axes.plot(x,y1, label = 'y1')
axes.plot(x,y2, label = 'y2')
plt.show()
```
