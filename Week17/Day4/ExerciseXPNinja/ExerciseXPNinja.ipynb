```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d as plt3d
import pandas as pd
```

```python
df = pd.DataFrame({
    'Temperature': [20, 22, 24, 26, 28],
    'Rainfall': [1, 2, 1, 0, 1],
    'Crop Yield': [100, 120, 140, 160, 180]
})

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Temperature'], df['Rainfall'], df['Crop Yield'])
ax.set_xlabel('Temperature')
ax.set_ylabel('Rainfall')
ax.set_title('3d scatter plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
embarked = titanic.groupby('embark_town').count()['embarked']
plt.pie(list(embarked.values), labels=list(embarked.index.values), autopct='%1.1f%%')
plt.show()
```
