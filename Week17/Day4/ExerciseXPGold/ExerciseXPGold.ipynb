```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
   'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
   'Temperature': [22, 24, 23, 24, 25, 26, 25],
   'Rainfall': [1, 0, 2, 1, 0, 0, 1]
})

day = list(df['Day'].values)
temp = list(df['Temperature'].values)
rainfall = list(df['Rainfall'].values)

plt.errorbar(day, temp, yerr=1)
plt.plot(day, temp, color='blue')
plt.errorbar(day, rainfall, yerr=1)
plt.plot(day, rainfall, color='orange')
plt.legend(['Temperature', 'Rainfall'])
```

```python
x = np.arange(0,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots()
ax1.set_xlabel('X')
ax1.set_ylabel('Y1 Data', color='blue')
ax1.plot(x,y1, color='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Y2 Data', color='green')
ax2.plot(x,y2, color='green')

plt.show()
```
